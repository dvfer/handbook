The idea is to code a attempt of a train loop of a linear regression with combinatorial optimization techniques

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.random.random(100)
y = np.random.random(100)

phi_i = np.random.random(2)
print(phi_i)
```

```python
phi_i[0] + phi_i[1] * x
```

## Hill climbing


```python
current = phi_i.copy()
criterion = lambda y_hat, y: np.mean(np.sqrt(np.power(y - y_hat, 2)))
current_prediction = current[0] + current[1] * x
best_loss = criterion(current_prediction, y)
losses = [best_loss]
for _ in range(1000):
    mutations = np.random.random(2)
    neighbors = [
        current + mutations,
        current - mutations,
        current / mutations,
        current * mutations,
        current ** mutations,
        mutations
    ]
    for neig in neighbors:
        y_hat = neig[0] + neig[1] * x
        loss_hat = criterion(y_hat, y)
        if loss_hat < best_loss:
            current = neig.copy()
            best_loss = loss_hat
            losses.append(loss_hat)
        
print(losses[-1])
```

```python

plt.plot(losses)
```

```python
x_line = np.linspace(min(x), max(x), 100)
y_line = current[0] + current[1] * x_line
plt.scatter(x,y )
plt.plot(x_line, y_line)
```
